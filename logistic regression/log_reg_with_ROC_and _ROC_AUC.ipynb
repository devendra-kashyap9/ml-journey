```python
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import roc_curve, roc_auc_score
from matplotlib import pyplot as plt
```

```python
# Generate 2 class dataset
X, y = make_classification(n_samples=1000,
                           n_classes=2,
                           random_state=1)
```

```python
# split the dataset into train test
X_train, X_test, y_train, y_test = train_test_split(X, y,
                                                    test_size=0.25,
                                                    random_state=0)
```

```python
# Generate a no skill prediction - (Majority class)
# will alwyas give "0"
dummy_model_prob = [0 for _ in range(len(y_test))]
dummy_model_prob
```

```python
# Basic Logistic model
model = LogisticRegression()
model.fit(X_train, y_train)
```

```python
# Prediction based on probabilities
model_prob = model.predict_proba(X_test)
```

```python
# Let's focus on Positive outcome i.e. 1
model_prob = model_prob[:, 1]
```

```python
# Calculate the score
dummy_model_auc = roc_auc_score(y_test, dummy_model_prob)
model_auc = roc_auc_score(y_test, model_prob)
print(f"{dummy_model_auc = }")
print(f'{model_auc = }')
```

ROC_AUC Curves gives us the two output, i.e. TPR(True Positive Rate), FPR(False Positive rate).

$$ TPR = TP / P = TP / (TP + FN) = 1 - FPR $$

plot `TPR in y-axis and FPR in x-axis`

```python
# Calculate ROC Curves
dummy_fpr, dummy_tpr, _ = roc_curve(y_test, dummy_model_prob)
model_fpr, model_tpr, thresholds = roc_curve(y_test, model_prob)
```

```python
# plot  ROC Curve for the model
plt.plot(dummy_fpr, dummy_tpr, 
         linestyle='--',
         label="Dummy Model")

plt.plot(model_fpr, model_tpr,
         marker='.',
         label="Logistic")

plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.legend()

plt.show()
```

```python
import numpy as np
plt.figure(figsize=(20,40))
plt.plot(model_fpr, model_tpr, 
         label="Logistic")

for fpr, tpr, thr in zip(model_fpr, model_tpr, thresholds):
    plt.annotate(
        str(round(thr, 2)),
        (fpr, tpr)
    )

plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.legend()
plt.show()
```
