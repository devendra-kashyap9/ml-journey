```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
from sklearn.datasets import make_classification
```

```python
# Create the dataset
X, y = make_classification(n_samples=1000,
                    n_features=10,
                    n_informative=3,
                    n_classes=3,
                    random_state=42)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y,
                                                    test_size=0.3,
                                                    random_state=42)
```

For multiclass problems (whenever n_classes >= 3), all solvers except 'liblinear' optimize the (penalized) multinomial loss. 'liblinear' only handles binary classification but can be extended to handle multiclass by using ~sklearn.multiclass.OneVsRestClassifier.

```python
from sklearn.linear_model import LogisticRegression
logistic = LogisticRegression()
logistic.fit(X_train, y_train)
y_pred = logistic.predict(X_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
score = accuracy_score(y_test, y_pred)
cm = confusion_matrix(y_test, y_pred)
classificn_report = classification_report(y_test, y_pred)

print(f"{score=}")
print(cm)
print(f'Classification report:\n {classificn_report}')
```
