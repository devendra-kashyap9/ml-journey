```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Lets create synthetic data points
from sklearn.datasets import make_classification
```

```python
X, y = make_classification(n_samples=1000,
                           n_features=2,
                           n_classes=2,
                           n_clusters_per_class=2,
                           n_redundant=0,
                           random_state=10)
```

```python
X
```

```python
y
```

```python
sns.scatterplot(x=pd.DataFrame(X)[0], 
                y=pd.DataFrame(X)[1] ,
                hue=y)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,
                                                    test_size=0.25,
                                                    random_state=10)
```

```python
from sklearn.svm import SVC
```

```python
svc = SVC(kernel='linear')
```

```python
svc.fit(X_train, y_train)
```

```python
svc.intercept_, svc.coef_
```

```python
# Prediction
y_pred = svc.predict(X_test)
y_pred
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
print(classification_report(y_test, y_pred))
```

```python
print(confusion_matrix(y_test, y_pred))
```

Tweak the earlier code with param `n_clusters_per_class` while creating the dataset with more than 1, and try with `linear` kernel, you will clearly observe the **droppong** of accuracy, because datapoints are overlapped and they are not linearly seperable.

### Let's try with different kernel

> **rbf**

```python
rbf = SVC(kernel='rbf')  # thogh rbf is default kernel
```

```python
rbf.fit(X_train, y_train)
```

```python
# Prediction
y_pred_rbf = rbf.predict(X_test)
y_pred_rbf
```

```python
print(classification_report(y_test, y_pred_rbf))
```

> **Polynomial Kernel**

```python
poly = SVC(kernel='poly')
poly.fit(X_train, y_train)

# Prediction
y_pred_poly = poly.predict(X_test)
print(classification_report(y_test, y_pred_poly))
print(confusion_matrix(y_test, y_pred_poly))
```

> **Sigmoid**

```python
sigmoid = SVC(kernel='sigmoid')
sigmoid.fit(X_train, y_train)

# Prediction
y_pred_sigmoid = sigmoid.predict(X_test)
print(classification_report(y_test, y_pred_sigmoid))
print(confusion_matrix(y_test, y_pred_sigmoid))
```

🔑 **Note**: Our `rbf` kernel better for this given dataset than any of the available kernel; so always check the dataset before applying the kernel function.  

### Hyperparameter Tuning

```python
from sklearn.model_selection import GridSearchCV

# defining parameter range
param_grid = {
    'C': [0.1, 1, 10, 100, 1000],
    'gamma': [1, 0.1, 0.01, 0.001, 0.0001],
    'kernel': ['rbf']
}
```

```python
grid = GridSearchCV(SVC(),
             param_grid=param_grid,
             refit=True,
             cv=5,
             verbose=3)
```

```python
grid.fit(X_train, y_train)
```

```python
grid.best_params_
```

```python
grid.best_score_
```

```python
# Predictions
y_pred_grid = grid.predict(X_test)
print(classification_report(y_test, y_pred_grid))
print(confusion_matrix(y_test, y_pred_grid))
```

### SVM Kernel in-depth Intution

> Creating datapoints

```python
import numpy as np
import matplotlib.pyplot as plt 

x = np.linspace(-5.0, 5.0, 100)
y = np.sqrt(10**2- x**2)
x = np.hstack([x, -x])
y = np.hstack([y, -y])
```

```python
x1 = np.linspace(-5.0, 5.0, 100)
y1 = np.sqrt(5**2- x1**2)
x1 = np.hstack([x1, -x1])
y1 = np.hstack([y1, -y1])
```

```python
plt.scatter(y, x)
plt.scatter(y1, x1);
```

```python
import pandas as pd

df1 = pd.DataFrame(np.vstack([y, x]).T,
                             columns=["X1", "X2"])
df1['Y'] = 0
df2 = pd.DataFrame(np.vstack([y1, x1]).T,
                             columns=["X1", "X2"])
df2['Y'] = 1
df = pd.concat([df1, df2])
df.head()
```

```python
df.tail()
```

```python
## segregating dependent and independent features
X = df.iloc[:, :2]
y = df.Y
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,
                                                    test_size=0.25,
                                                    random_state=10)
```

```python
y_train
```

```python
# we need to find components for the Polynomial kernel
# X1, X2, X1-square, X2_square, X1*X2
df['X1_Square'] = df['X1'] ** 2
df['X2_Square'] = df['X2'] ** 2
df['X1*X2'] = df.X1 * df.X2
df.head()
```

```python
# Indepoendent and dependent features
X = df.drop("Y", axis=1)
y = df["Y"]
```

```python
X.columns, y.name
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,
                                                    test_size=0.25,
                                                    random_state=10)
```

```python
X_train
```

```python
# plotting 
import plotly.express as px

fig = px.scatter_3d(df, x='X1', y='X2', z='X1*X2',
                    color='Y')
fig.show()
```

```python
fig = px.scatter_3d(df, x='X1_Square', y='X2_Square', z='X1*X2',
                    color='Y')
fig.show()
```

### 📌 X1, X2 were not clearly separable but after squaring it become clearly separable.
