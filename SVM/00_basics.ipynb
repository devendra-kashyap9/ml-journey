### Support Vector Machine 

* Support Vector Regressor
* Support Vector Classifier

- Marginal Plane and best fit line
 Increase the distance between plane and line, and they should be close to the data points.

 Main Aim to to find the best fit line in Classifer problem with maximum distance between plane and line.

 * Soft Margin - some amount of error will be there i.e. overlapping points.
 * Hard Margin - clear seperation.

* Hinge Loss - A loss function primarily used for training maximum margin classifier, most notably SVM. It penalize the data points that are either **misclassified or coreectly classified** but too close to *decision boundry*.

**C** a regularization parameter balancing classification confidence with margin strictness.

> **C** is inversely proportion to **Cost Function**
