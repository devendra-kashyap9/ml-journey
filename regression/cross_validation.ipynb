### cross Val technique

* Leave One Out CV(LOOCV)
        one record as validation data and rest as training data
        
        cons: 
            - complexity of training the model increases cuz we are taking one model
            - may ead to overfit (high training data) less validation data

* Leave P out CV
        instead of one some p size is choosen

* K-fold Cross validation
        k=5, n= 500
        test_size = n/5 -> 100 = validation and rest are training data
        and accuracy will be avg of all

        cons:
            - not able to shuffle so chances are not getting proper mixing or all combination

* Stratified K-fold cross validation
        stratify = label
        so validation datatset will be in proportion to the the labels(outputs)

* Time Series cross validation
        reviews and time 
        cv based on time/period
