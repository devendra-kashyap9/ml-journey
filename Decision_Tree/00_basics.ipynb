### Decision Tree Classifier: In-Depth Machine Learning Theory

A **Decision Tree** is a non-parametric supervised learning algorithm used for both classification and regression tasks. It has a hierarchical, tree-like structure consisting of root nodes, branches, internal nodes, and leaf nodes. It mimics human decision-making by breaking down a complex dataset into smaller, manageable subsets based on conditional rules.

---

#### 1. Structural Anatomy of a Decision Tree

Before looking into the math, it helps to understand the foundational anatomy of the tree:

* **Root Node:** The topmost node that represents the entire dataset. This is where the first split occurs.
* **Internal / Decision Node:** Nodes where the data splits into sub-nodes based on a specific feature condition.
* **Leaf / Terminal Node:** Endpoints of the tree that do not split further. They represent the final class label or predicted value.
* **Branches:** The paths connecting nodes, representing the outcome of a decision rule (e.g., `True` or `False`).



---

#### 2. Mathematical Fundamentals: How a Tree Splits

A decision tree splits nodes in a way that maximizes the **homogeneity** (purity) of the resulting child nodes. It determines the best feature to split on using specific mathematical metrics.

##### A. Entropy and Information Gain (Mainly used in ID3 / C4.5 algorithms)

**Entropy** measures the degree of randomness, disorder, or impurity in a dataset. It ranges from $0$ (perfectly pure) to $1$ (completely random/equal distribution).

For a dataset $S$ with $c$ classes, the Entropy $H(S)$ is calculated as:

$$H(S) = -\sum_{i=1}^{c} p_i \log_2(p_i)$$

Where $p_i$ is the probability/proportion of samples belonging to class $i$.

**Information Gain (IG)** measures the reduction in entropy after a dataset is split on a specific feature $A$. The tree chooses the feature that yields the highest Information Gain.

$$IG(S, A) = H(S) - \sum_{v \in \text{Values}(A)} \frac{|S_v|}{|S|} H(S_v)$$

---

##### B. Gini Impurity (Default metric used in Scikit-Learn's CART(Classification and Regression Tree) algorithm)

**Gini Impurity** measures the probability of a randomly chosen element from the dataset being incorrectly labeled if it was randomly labeled according to the distribution of labels in the subset. It ranges from $0$ (pure) to $0.5$ (equally distributed binary classes).

$$Gini(S) = 1 - \sum_{i=1}^{c} (p_i)^2$$

Like Information Gain, the algorithm calculates the **Gini Gain** (reduction in Gini impurity) for all possible splits and chooses the split that minimizes the resulting child node impurities.

| Metric | Calculation Cost | Range | Behavior |
| :--- | :--- | :--- | :--- |
| **Gini Impurity** | Faster (no logarithmic calculations) | $0.0$ to $0.5$ (Binary) | Tends to isolate the largest class into its own branch. |
| **Entropy** | Slower (requires computing $\log_2$) | $0.0$ to $1.0$ | Tends to create better-balanced tree structures. |

---

#### 3. The Overfitting Dilemma and Regularization

Because decision trees can keep splitting until every single training observation sits in its own pure leaf node, they are highly prone to **overfitting**. An overfitted tree memorizes the noise in the training data and fails to generalize to test data.



To prevent this, we regularize the tree using **Hyperparameter Tuning** or **Pruning**:

* **`max_depth`**: Restricts how deep the tree can grow. Lower numbers prevent overfitting.
* **`min_samples_split`**: The minimum number of samples a node must contain before it can be split.
* **`min_samples_leaf`**: The minimum number of samples allowed inside a terminal leaf node.
* **`max_features`**: Limits the number of features evaluated when looking for the best split.

---

#### 4. Pros and Cons

##### **The Good**
* **White-Box Model:** Highly interpretable. You can visually trace exactly *why* a model made a specific prediction.
* **No Feature Scaling Required:** Monotonic transformations do not affect splits; hence, normalization or standardization is unnecessary.
* **Handles Mixed Data Types:** Naturally handles both categorical and continuous features out of the box.

##### **The Bad**
* **Highly Unstable:** A minor change in the training dataset can result in a completely different tree structure (High Variance).
* **Greedy Approach:** It makes locally optimal decisions at each node rather than looking ahead to find a globally optimal tree.
* **Biased Toward Dominant Classes:** If some classes dominate the dataset, the tree will build rules biased toward them (requires balancing).



**Use Case**

- for small data use `entropy` and for larger data set `gini`.

### Pruning in Decision Tree

Pruning is a technique used to reduced `Overfitting`.

* Post Pruning:
    * Construct DT
    * Prune it wrt `depth`
    * Applicable for smaller dataset

* Pre Pruning:
    * We `tweak/tune hyperparameter` while constructing DT.
