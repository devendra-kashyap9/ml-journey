* `Variance reduction` is a technique we use for Information gain in regression.

$$ {VarianceReduction} = {VarianceOfRoot} - \sum{wi * Variance(Child)}$$

> Lower the value better the split.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 

%matplotlib inline
```

```python
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
```

```python
iris
```

```python
print(iris.DESCR)
```

```python
# independent features
X = pd.DataFrame(iris.data,
                 columns=['sepal length', 'sepal width', 'petal length', 'petal width'])
```

```python
# dependent features
y = iris.target
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y,
                                                    test_size=0.2,
                                                    random_state=10)
```

```python
# apply decision tree classifier
from sklearn.tree import DecisionTreeClassifier
treeclassifier = DecisionTreeClassifier()
treeclassifier.fit(X_train, y_train)
```

```python
# visualize decisiom tree
from sklearn import tree
plt.figure(figsize=(15,10))
tree.plot_tree(treeclassifier, filled=True);
```

### Post Pruning

Lets cut the branches after `depth=2`

```python
treeclassifier = DecisionTreeClassifier(max_depth=2)
treeclassifier.fit(X_train, y_train)
```

```python
# visualize decisiom tree
from sklearn import tree
plt.figure(figsize=(15,10))
tree.plot_tree(treeclassifier, filled=True);
```

```python
y_pred = treeclassifier.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix, classification_report, ConfusionMatrixDisplay
```

```python
cm = confusion_matrix(y_test, y_pred)
print(cm)

ConfusionMatrixDisplay.from_predictions(y_test, y_pred);
```

```python
print(classification_report(y_test, y_pred))
```

### Pre Pruning and Hyperparameter Tuning

```python
params = {
    'criterion': ['gini', 'entropy', 'log_loss'],
    'splitter': ['best', 'random'],
    'max_depth': [1,2,3,4,5],
    'max_features': ['auto', 'sqrt', 'log2']
}

params
```
