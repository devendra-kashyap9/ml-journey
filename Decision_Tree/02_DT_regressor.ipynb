```python
# Load the diabetes dataset
from sklearn.datasets import load_diabetes

diabetes = load_diabetes()
```

```python
diabetes
```

```python
# description of dataset
print(diabetes.DESCR)
```

```python
import pandas as pd
df_diabetes = pd.DataFrame(diabetes.data,
                           columns=['age', 'sex', 'bmi', 'bp', 's1', 's2', 's3', 's4', 's5', 's6'])

df_diabetes.head()
```

```python
# Independent and dependent feature
X = df_diabetes
y = diabetes.target
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y,
                                                    test_size=0.3,
                                                    random_state=10)
```

```python
X_train.corr() # default is 'pearson'
```

```python
import seaborn as sns 
import matplotlib.pyplot as plt
%matplotlib inline

plt.figure(figsize=(15, 10))
sns.heatmap(X_train.corr(), annot=True);
```

```python
from sklearn.tree import DecisionTreeRegressor

regressor = DecisionTreeRegressor()
regressor.fit(X_train, y_train)
```

### Hyperoparmater Tunning

```python
params = {
    'criterion': ['squared_error', 'friedman_mse', 'absolute_error', 'poisson'],
    'splitter': ['best', 'random'],
    'max_depth': [1,2,3,4,5,10,15,20,25],
    'max_features': ['auto', 'sqrt', 'log2']
}
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
grid = GridSearchCV(estimator=regressor,
                    param_grid=params,
                    cv=5,
                    scoring='neg_mean_squared_error')
```

```python
import warnings
warnings.filterwarnings('ignore')

grid.fit(X_train, y_train)
```

```python
grid.best_params_
```

```python
# Prediction
y_pred_grid = grid.predict(X_test)
```

```python
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error

print(r2_score(y_test, y_pred_grid))
print(mean_absolute_error(y_test, y_pred_grid))
print(mean_squared_error(y_test, y_pred_grid))
```

### Deciiosn Tree Plotting

```python
selected_model = DecisionTreeRegressor(criterion='poisson',
                                       max_depth=4,
                                       max_features='sqrt',
                                       splitter='best')

selected_model.fit(X_train, y_train)
```

```python
from sklearn import tree

plt.figure(figsize=(15, 10))
tree.plot_tree(selected_model, filled=True);
```
